# DATA PREPROCESSING AND EDA

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop('customerID',axis=1,inplace=True)
```

```python
df.head()
```

### changing object data type to numeric data type

```python
df.columns
```

```python
binary_cols=['gender','Partner','Dependents','PhoneService','PaperlessBilling','Churn']
```

```python
df[binary_cols]=df[binary_cols].replace({
    'Yes':1,'No':0,
    'Male':1,'Female':0})
```

```python
df.info()
```

```python
multi_cols=['MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV',
            'StreamingMovies','Contract','PaymentMethod']         
```

```python
df=pd.get_dummies(df,columns=multi_cols,drop_first=True)
```

```python
df.info()
```

```python
bool_cols=df.select_dtypes(include='bool').columns
df[bool_cols]=df[bool_cols].astype(int)
```

```python
df.info()
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df.info()
```

### Correlation check with respect to Churn Label

```python
corr_matrix=df.corr(numeric_only=True)
churn_corr=corr_matrix[['Churn']].sort_values(by='Churn',ascending=False)
```

```python
plt.figure(figsize=(4,12))
sns.heatmap(churn_corr,annot=True,cmap='coolwarm',vmin=1,vmax=1)
plt.title('Correlation with respect to Churn')
plt.show();
```

# MODEL TRAINING AND EVALUATIONS

```python
from sklearn.model_selection import train_test_split
```

```python
df=df.dropna()
X=df.drop('Churn',axis=1)
y=df['Churn']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

### TRAINING USING XGBOOST CLASSIFIER

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
pred=xgb.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

### EVALUATION FOR XGBOOST CLASSIFIER

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
y.value_counts(normalize=True).plot(kind='bar')
plt.show();
```

### Because the data is imbalanced,we will use different techniques to improve the recall(for 1) in this case

## XGBOOST + THRESHOLD

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
proba=xgb.predict_proba(X_test)[:,1]
```

```python
print("Threshold tuning for XGBoost")
print(f"{'Thresh':<8}{'prec':<8}{'recall':<8}{'F1':<8}")
for thresh in [0.25,0.30,0.35,0.40,0.45,0.50]:
    preds=(proba>= thresh).astype(int)
    prec=precision_score(y_test,preds,pos_label=1)
    rec=recall_score(y_test,preds,pos_label=1)
    F1=f1_score(y_test,preds,pos_label=1)
    print(f"{thresh:<8}{prec:<8.3f}{rec:<8.3f}{F1:<8.3f}")

 
```

#### using 0.25 as the new threshold

```python
new_thresh=0.25
proba=xgb.predict_proba(X_test)[:,1]
y_pred=(proba>=new_thresh).astype(int)
```

## EVALUATIONS FOR XGBOOST + THRESHOLD

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

### XGBOOST + SMOTE 

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(random_state=42)
X_train_sm,y_train_sm=sm.fit_resample(X_train,y_train)
xgb_sm=XGBClassifier(random_state=42)
xgb_sm.fit(X_train_sm,y_train_sm)
y_pred_sm=xgb_sm.predict(X_test)

```

## EVALUATIONS FOR XGBOOST + SMOTE

```python
print("XGBOOST + SMOTE RESULTS")
print(classification_report(y_test,y_pred_sm))
print(confusion_matrix(y_test,y_pred_sm))
```
